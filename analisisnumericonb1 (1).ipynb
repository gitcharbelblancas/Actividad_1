```
print ('Problema 4')
print ('Menu\t'

  'a) Suma\t'
  'b) Resta\t'
  'c) Multiplicacion\t'
  'd) División\t')

def Calculadora():
  opc=int(input('Seleccione Opcion'))
  while (opc >0 and opc <4):
    a=int(input('De un número'))
    b=int(input('De otro número'))
  
    if(opc==1):
       print ('La suma es:')
       print (a+b)
      
    elif opc==2:
       print ('La resta es:')
       print (a-b)
       
    elif opc==3:
       print ('El producto es:')
       print (a*b)
       return 

    elif opc==4: 
       print ('El cociente es:')
       print (a/b)   
    while b==0:
       return [0]

calculadora() 

      







```

```
print ('Problema 8')
def triangulo(n):
    return "\n".join(["*"*i for i in range(1,n+1)])
numero=int(input("indica un numero: "))
print(triangulo(numero))
```
